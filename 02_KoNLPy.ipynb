## 형태소 분석기 공통 메소드
- `morphs(string)` : 형태소 단위로 토큰화(tokenize)
- `nouns(string)` : 명사만 추출하여 토큰화(tokenize)    
- `pos(string)`: 품사 부착
    - 형태소 분석기마다 사용하는 품사태그 상이
        - https://konlpy-ko.readthedocs.io/ko/v0.5.2/morph/
- `tagset`: 형태소 분석기가 사용하는 품사태그 설명하는 속성

```python
from konlpy.tag import Okt, Komoran    # Okt: Open Korea Tet
```

```python
txt = """1인칭 슈팅게임(FPS) ‘크로스파이어’ 제작사인 스마일게이트가 지난해 매출 1조원을 넘겼다. 국내 게임사 중 다섯 번째로 1조원 이상의 매출을 기록했다.

13일 스마일게이트 감사보고서에 따르면, 이 회사는 지난해 연결 기준 매출 1조 73억원, 영업이익 3646억원을 기록했다. 직전년도 대비 매출은 14%, 영업이익은 26% 늘어난 수치다. 회사 관계자는 “창사 이후 18년 만에 처음으로 매출 1조원을 넘겼다”고 설명했다.

실적 상승을 견인한 건 해외 매출이다. 스마일게이트가 지난해 해외에서 벌어들인 돈은 8430억원에 달한다. 전체 매출의 83.7%다. 주력 지식재산(IP)인 ‘크로스파이어’의 역할이 컸다. 이 게임은 2008년 중국 서비스를 시작한 이후 글로벌 시장에서 큰 성과를 냈다. 2017년엔 글로벌 동시 접속자 수 800만명을 넘겼다. 글로벌 이용자 수는 6억명 이상(2020년)이다. 문화체육관광부는 지난해 문화 수출 공로를 인정해 창업자인 권혁빈 스마일게이트 희망스튜디오 이사장에게 보관문화훈장을 수여했다.

스마일게이트는 앞으로 ‘콘솔게임’ 시장에 본격 진출할 계획이다. 크로스파이어 IP를 활용한 콘솔게임 ‘크로스파이어X’를 올해 마이크로소프트와 함께 선보인다. 스마일게이트 홀딩스 성준호 대표는 “크로스파이어X를 시작으로, 한 차원 높은 수준의 작품으로 새로운 도전을 이어가겠다”고 말했다.

스마일게이트를 포함해 지난해 매출 1조원 이상을 기록한 국내 게임사는 총 5곳이다. 넥슨이 3조 1306억 원으로 매출 1위에 올랐고, 넷마블(2조 4848억원) · 엔씨소프트(2조 4162억원) · 크래프톤(1조 6704억원) 순으로 뒤를 이었다. 게임 이외 사업 매출이 더 큰 NHN은 제외했다.
"""
```

# KoNLPy 형태소 분석기와 NLTK를 이용한 문서의 분석

```python
from nltk import Text
from konlpy.corpus import kolaw
from konlpy.tag import Komoran
```

```python
law = kolaw.open('constitution.txt').read()
print(law[:100])
```

```python
# 명사만 추출 - 특수문자들도 빠짐
kom = Komoran()
nouns = kom.nouns(law)
len(nouns)
```

```python
# 1 글자인 명사들은 제거
nouns = [word for word in nouns if len(word) > 1]
len(nouns)
```

```python
text = Text(nouns, name = '헌법')
text
```

```python
"대한민국 빈도수: {}, 헌법 빈도수: {}, 국민 빈도수: {}".format(text.count('대한민국'), text.count('헌법'), text.count('국민'))
```

```python
# matplotlib 한글 처리
import matplotlib as mpl
mpl.rcParams['font.family'] = 'malgun gothic'
```

```python
 # 빈도수 상위 20개로 그래프 그리기
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 7))

plt.title('헌법에 나온 명사 빈도수 상위 20개', color = 'w')
text.plot(20)

plt.show()
```

```python
plt.figure(figsize = (15, 10))
text.dispersion_plot(['법률', '대통령', '국가', '국회', '헌법'])
```

```python
# 매개변수로 전달한 단어가 나오는 라인을 출력
# width - 국민을 기준으로 50글자(width 수만큼) 보여줌
# line - 조회할 줄의 개수 
# 67개 중에 10개만 보여줌
text.concordance('국민', width = 50, lines = 10)
```

```python
# 단어별 빈도수 - freqdist
freq = text.vocab()
freq
```

```python
f"총 단어수: {freq.N()}, 고유단어: {freq.B()}, 최빈단어: {freq.max()}, 최빈단어의 빈도수: {freq.get(freq.max())}, 최빈단어의 출연비율: {freq.freq(freq.max())}"
```

```python
word_cnt = freq.most_common()
word_cnt
```

```python
import pandas as pd
df = pd.DataFrame(word_cnt, columns=['단어', '빈도수'])
df.head(10)
```

```python
# 빈도수 = 1인 단어들의 경우, 분석할 때 제외할 수 있음
df.tail(10)
```

```python
# 5번 이상 나온 단어들만 조회
df2 = df[df['빈도수'] >= 5]
df2.tail()
```
